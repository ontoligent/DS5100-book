# An alternate solution to HW03 Task 1

The main ideas are:
1. Expressions behave as values.
2. Boolean expressions evaluate to numeric values.
3. Compound expressions can be replaced by arithmetic.

Regarding #3, note how conditions 1, 2, and 3 are met implicitly by the use of multiplication and addition.

````python
N = range(1, 101)
A = [(x % 3 == 0) * 'Wahoo' for x in N]
B = [(x % 5 == 0) * 'wah!' for x in N]
C = [a + b for a, b in zip(A, B) if a + b]
````

````python
print('\n'.join(C))
print(len(C))
````

# In-Place Functions

In-place functions directly change the value of what they operate on,\
as opposed to expecting these changes to be returned and used to update a value.

For example, a typical function may work like this:

```python
x = 1 
add_one(x)
# x remains the same.

x = add_one(x)
# x is now 2
```

A function that performs an in-place operation would work like this:

```python
x = 1
add_one(x)
# x is now 2
```

````python
x = [i for i in range(10)]
````

# In-Place Operations

Python has a library of functions that perform in-place operations.

`a += b` is the same as `operator.iadd(a, b)`

````python
import operator
````

````python
a = [1,2,3]
b = [4,5,6]
````

````python
operator.iadd(a, b)
````
